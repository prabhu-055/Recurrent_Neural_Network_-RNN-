```python
# Step 1
# Import necessary libraries
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense
```

```python
# Step 2
# Generate training data: sequences like [1, 2, 3] -> 4
X = []
y = []
for i in range(1, 10):
  X.append([i, i+1, i+2])  # input sequence
  y.append(i+3)           # next number

# Convert to numpy arrays and reshape for RNN (samples, timesteps, features)
X = np.array(X).reshape(len(X), 3, 1)
y = np.array(y)
```

```python
# Build the RNN model
model = Sequential()
model.add(SimpleRNN(10, activation='relu', input_shape=(3, 1))) # 10 RNN units
model.add(Dense(1)) # Output layer with 1 neuron
model.compile(optimizer='adam', loss='mse') # Compile using Adam optimizer and Mean Squared Error loss
```

```python
# Step 4
# Train the model
model.fit(X, y, epochs=300, verbose=1)
```

```python
# Step 4
# Test the model with a new sequence
test_input = np.array([[10, 11, 12]]).reshape((1, 3, 1))
predicted = model.predict(test_input, verbose=0)

print(f"Input Sequence: [15, 80, 32]")
# Note: Accessing the prediction might vary slightly depending on TF/Keras version
# This line assumes the prediction is nested; adjust if necessary.
print(f"Predicted Next Number: {round(float(predicted[0][0]), 2)}")
```
